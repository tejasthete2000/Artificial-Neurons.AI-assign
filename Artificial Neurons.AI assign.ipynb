# @@@ create a function to identify weak area and averange area and strong area as par unit sealing @@@

```python
df.head(2)
```

```python
def descripation(ab):
    if (ab>0) and (ab<2):
        range1=('poor_area')
    elif (ab>2) and (ab<4):
        range1=('Medium_area')
    else:
        range1=('strong_area')
    return range1
```

```python
df['descripation_range']=df['UNIT'].apply(descripation)
```
